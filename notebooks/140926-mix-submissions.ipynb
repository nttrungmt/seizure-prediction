```python
df = pd.DataFrame()
df['gb'] = pd.read_csv('../submissions/140926-predict.8.csv', index_col='clip', squeeze=True) #64
df['rf'] = pd.read_csv('../submissions/140926-predict.2.csv', index_col='clip', squeeze=True)
```

```python
pd.scatter_matrix(df[['gb','rf']],figsize=(6, 6), diagonal='kde');
```

```python
w_gb  = 0.2
w_rf = 1

s = w_gb + w_rf

w_gb /= s
w_rf /= s
```

```python
w_gb, w_rf
```

```python
df['preictal'] = w_gb * df['gb'] + w_rf * df['rf']
```

```python
df['preictal'].to_csv('../submissions/140926-predict.9.csv', header=True)
```
